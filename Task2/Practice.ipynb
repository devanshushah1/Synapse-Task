```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

data = pd.read_json('data.json')
total = pd.json_normalize(data['data'])
```

```python
total.head()
```

```python
cleaned = total[(total.category == "Sports")|(total.category == "Environment")]
cleaned
```

```python
funded = cleaned[cleaned.event_name == "Fund Project"]
funded
```

```python
plt.style.use('fivethirtyeight')
latitude = [round(x) for x in cleaned['location.latitude']]
longitude = [round(x) for x in cleaned['location.longitude']]
plt.scatter(latitude, longitude, alpha=0.3)
plt.ylabel("longitude")
plt.xlabel("latitude")
plt.show()
```

```python
plt.style.use('seaborn')
m = funded['marital_status'].value_counts()
m.plot(kind='pie', autopct = '%.2f', labels = ['married', 'single'])
plt.ylabel('')
plt.show()
```

```python
plt.style.use('seaborn')
m = funded['device'].value_counts()
m.plot(kind='pie', autopct = '%.2f', labels = ['ios', 'android'])
plt.ylabel('')
plt.show()
```
